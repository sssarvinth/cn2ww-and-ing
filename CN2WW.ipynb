```python
##convert all the dates to datetime types, and handle the column values not inlined with the format


df['Pickupdate\n揽收日期']=pd.to_datetime(df['Pickupdate\n揽收日期'], errors='coerce')
df['Shipping_Out_from_Origin_Date\n离港日期']=pd.to_datetime(df['Shipping_Out_from_Origin_Date\n离港日期'], errors='coerce')
df['Arrival at destinaton country\n抵港日期']=pd.to_datetime(df['Arrival at destinaton country\n抵港日期'], errors='coerce')
df['Custom_clearance_date\n清关日期']=pd.to_datetime(df['Custom_clearance_date\n清关日期'], errors='coerce')
df['FC_received_date\n入仓日期']=pd.to_datetime(df['FC_received_date\n入仓日期'], errors='coerce')

## hand double spacing between words
df=df.replace('\s+', ' ', regex=True)
```

```python
## handle datetime value and convert to y-m-d format

df['Pickupdate\n揽收日期']=df['Pickupdate\n揽收日期'].dt.strftime("%Y-%m-%d")
df['Shipping_Out_from_Origin_Date\n离港日期']=df['Shipping_Out_from_Origin_Date\n离港日期'].dt.strftime("%Y-%m-%d")
df['Arrival at destinaton country\n抵港日期']=df['Arrival at destinaton country\n抵港日期'].dt.strftime("%Y-%m-%d")
df['Custom_clearance_date\n清关日期']=df['Custom_clearance_date\n清关日期'].dt.strftime("%Y-%m-%d")
df['FC_received_date\n入仓日期']=df['FC_received_date\n入仓日期'].dt.strftime("%Y-%m-%d")
```

```python
##handle the spaces before and after the string

df['Carrier_Name\n服务商名称']=df['Carrier_Name\n服务商名称'].str.strip()
#df['Service_provider_ID(by Amazon)\n服务商ID（亚马逊提供）']=df['Service_provider_ID(by Amazon)\n服务商ID（亚马逊提供）'].str.strip() 
df['Type of service\n运输方式']=df['Type of service\n运输方式'].str.strip() 
df['Last mile solution (Only CN2IN)\n送仓方式(适用于印度站)']=df['Last mile solution (Only CN2IN)\n送仓方式(适用于印度站)'].str.strip()
df['Shipment_id\nFBA单号']=df['Shipment_id\nFBA单号'].str.strip()
df['Seller Name\n卖家名称']=df['Seller Name\n卖家名称'].str.strip()
df['Origin_port\n始发港']=df['Origin_port\n始发港'].str.strip()
df['Origin_country\n始发国家']=df['Origin_country\n始发国家'].str.strip()
df['Destination_port\n目的港']=df['Destination_port\n目的港'].str.strip()
df['Destination_country\n目的国']=df['Destination_country\n目的国'].str.strip()
df['Pickupdate\n揽收日期']=df['Pickupdate\n揽收日期'].str.strip()
df['Shipping_Out_from_Origin_Date\n离港日期']=df['Shipping_Out_from_Origin_Date\n离港日期'].str.strip()
df['Arrival at destinaton country\n抵港日期']=df['Arrival at destinaton country\n抵港日期'].str.strip()
df['Custom_clearance_date\n清关日期']=df['Custom_clearance_date\n清关日期'].str.strip()
df['FC_received_date\n入仓日期']=df['FC_received_date\n入仓日期'].str.strip()
df['Total_shipment_units\n入仓件数']=df['Total_shipment_units\n入仓件数'].str.strip() 
df['total_shipment_kgs\n货物重量(kg)']=df['total_shipment_kgs\n货物重量(kg)'].astype(str).str.strip()
df['total_shipment_kgs\n货物重量(kg)']=df['total_shipment_kgs\n货物重量(kg)'].astype(float)
df['total_shipment_volumn\n货物体积(cbm,适用海运)']=df['total_shipment_volumn\n货物体积(cbm,适用海运)'].str.strip() 
df['shipment_invoice_value\n申报金额(RMB)']=df['shipment_invoice_value\n申报金额(RMB)'].astype(str).str.strip() 
df['shipment_invoice_value\n申报金额(RMB)']=df['shipment_invoice_value\n申报金额(RMB)'].astype(float)
```
