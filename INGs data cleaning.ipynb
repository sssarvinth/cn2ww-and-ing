```python
## sellercustomerid should not be blank, replace the NaN with Null49

df['sellercustomerid'] = df['sellercustomerid'].fillna('NULL49')


## limit the port_to_door_of_reason to 100 columns
df['port_to_door_of_reason']=df['port_to_door_of_reason'].str[0:100]

## hand double spacing between words

df=df.replace('\s+', ' ', regex=True) 
```

```python
##convert all the dates to datetime types, and handle the column values not inlined with the format

df['FC_Appointment_received_Date']=pd.to_datetime(df['FC_Appointment_received_Date'], errors='coerce')


df['pickupdateinutc']=pd.to_datetime(df['pickupdateinutc'], errors='coerce')
df['deliverydateinutc']=pd.to_datetime(df['deliverydateinutc'], errors='coerce')
df['custom_clearance_date_of_origin']=pd.to_datetime(df['custom_clearance_date_of_origin'], errors='coerce')
df['Expected_Shipping_Out_from_Origin_Date']=pd.to_datetime(df['Expected_Shipping_Out_from_Origin_Date'], errors='coerce')
df['Actual_Shipping_Out_from_Origin_Date']=pd.to_datetime(df['Actual_Shipping_Out_from_Origin_Date'], errors='coerce')
df['Expected_Shipping_To_Transit_Date']=pd.to_datetime(df['Expected_Shipping_To_Transit_Date'], errors='coerce')
df['Actual_Shipping_To_Transit_Date']=pd.to_datetime(df['Actual_Shipping_To_Transit_Date'], errors='coerce')
df['Expected_Shipping_From_Transit_Date']=pd.to_datetime(df['Expected_Shipping_From_Transit_Date'], errors='coerce')
df['Actual_Shipping_From_Transit_Date']=pd.to_datetime(df['Actual_Shipping_From_Transit_Date'], errors='coerce')
df['Expected_Shipping_To_Destination_Date']=pd.to_datetime(df['Expected_Shipping_To_Destination_Date'], errors='coerce')
df['Actual_Shipping_To_Destination_Date']=pd.to_datetime(df['Actual_Shipping_To_Destination_Date'], errors='coerce')
df['FC_Appointment_received_Date']=pd.to_datetime(df['FC_Appointment_received_Date'], errors='coerce')


```

```python
## handle datetime value and convert to y-m-d format

df['pickupdateinutc']=df['pickupdateinutc'].dt.strftime("%Y-%m-%d")
df['deliverydateinutc']=df['deliverydateinutc'].dt.strftime("%Y-%m-%d")
df['custom_clearance_date_of_origin']=df['custom_clearance_date_of_origin'].dt.strftime("%Y-%m-%d")
df['Expected_Shipping_Out_from_Origin_Date']=df['Expected_Shipping_Out_from_Origin_Date'].dt.strftime("%Y-%m-%d")
df['Actual_Shipping_Out_from_Origin_Date']=df['Actual_Shipping_Out_from_Origin_Date'].dt.strftime("%Y-%m-%d")
df['Expected_Shipping_To_Transit_Date']=df['Expected_Shipping_To_Transit_Date'].dt.strftime("%Y-%m-%d")
df['Actual_Shipping_To_Transit_Date']=df['Actual_Shipping_To_Transit_Date'].dt.strftime("%Y-%m-%d")
df['Expected_Shipping_From_Transit_Date']=df['Expected_Shipping_From_Transit_Date'].dt.strftime("%Y-%m-%d")
df['Actual_Shipping_From_Transit_Date']=df['Actual_Shipping_From_Transit_Date'].dt.strftime("%Y-%m-%d")
df['Expected_Shipping_To_Destination_Date']=df['Expected_Shipping_To_Destination_Date'].dt.strftime("%Y-%m-%d")
df['Actual_Shipping_To_Destination_Date']=df['Actual_Shipping_To_Destination_Date'].dt.strftime("%Y-%m-%d")
df['FC_Appointment_received_Date']=df['FC_Appointment_received_Date'].dt.strftime("%Y-%m-%d")
```

```python
##handle the spaces

df['sellercustomerid']=df['sellercustomerid'].str.strip()
df['origin_port']=df['origin_port'].str.strip() 
df['destination_port']=df['destination_port'].str.strip() 
df['pickupdateinutc']=df['pickupdateinutc'].str.strip()
df['deliverydateinutc']=df['deliverydateinutc'].str.strip()
df['custom_clearance_date_of_origin']=df['custom_clearance_date_of_origin'].str.strip()
df['Expected_Shipping_Out_from_Origin_Date']=df['Expected_Shipping_Out_from_Origin_Date'].str.strip()
df['Actual_Shipping_Out_from_Origin_Date']=df['Actual_Shipping_Out_from_Origin_Date'].str.strip()
df['Expected_Shipping_To_Transit_Date']=df['Expected_Shipping_To_Transit_Date'].str.strip()
df['Actual_Shipping_To_Transit_Date']=df['Actual_Shipping_To_Transit_Date'].str.strip()
df['Expected_Shipping_From_Transit_Date']=df['Expected_Shipping_From_Transit_Date'].str.strip()
df['Actual_Shipping_From_Transit_Date']=df['Actual_Shipping_From_Transit_Date'].str.strip()
df['Expected_Shipping_To_Destination_Date']=df['Expected_Shipping_To_Destination_Date'].str.strip()
df['Actual_Shipping_To_Destination_Date']=df['Actual_Shipping_To_Destination_Date'].str.strip()
df['port_to_port_of_Reason']=df['port_to_port_of_Reason'].str.strip()
df['FC_Appointment_received_Date']=df['FC_Appointment_received_Date'].str.strip()  
df['liner_details']=df['liner_details'].str.strip()
df['port_to_door_of_reason']=df['port_to_door_of_reason'].str.strip()
df['Container #']=df['Container #'].str.strip()
df['Console #']=df['Console #'].str.strip()
df['Gross weight']=df['Gross weight'].astype(str).str.strip()
df['Gross weight']=df['Gross weight'].astype(float) # used strip to remove space in df, converted to float->str and float
df['total_shipment_volumn']=df['total_shipment_volumn'].astype(str).str.strip()
df['total_shipment_volumn']=df['total_shipment_volumn'].astype(float)
df['shipment_invoice_value']=df['shipment_invoice_value'].astype(str)
df['shipment_invoice_value']=df['shipment_invoice_value'].astype(float)
df['shipmentid']=df['shipmentid'].str.strip()
df['tracking_id']=df['tracking_id'].str.strip()  
df['us_coast_port']=df['us_coast_port'].str.strip() 
df['type_of_shipping_mode_offered']=df['type_of_shipping_mode_offered'].str.strip()
df['provider_id']=df['provider_id'].str.strip() 
df['reason_for_delay_at_origin']=df['reason_for_delay_at_origin'].str.strip()
df['reason_for_delay_at_destination_port']=df['reason_for_delay_at_destination_port'].str.strip()
df['reason_for_delay_at_destination_fc']=df['reason_for_delay_at_destination_fc'].str.strip()
df['reason_for_delay_at_transit_port']=df['reason_for_delay_at_transit_port'].str.strip()
```
